```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
df = pd.read_csv("merged.csv")
```

```python
df2020 = df.loc[df["year"]== 2020]
df2021 = df.loc[df["year"]== 2021]
df2021.head()
```

```python

df2021_2 = df2021.dropna(subset=['inflation', 'EconomicSupportIndex'])
df2021_2.head()
```

```python
slope, intercept, r, p, err = st.linregress(df2021_2["EconomicSupportIndex"],df2021_2["inflation"])
model = slope * df2021_2["EconomicSupportIndex"] + intercept

plt.scatter(df2021_2["EconomicSupportIndex"],df2021_2["inflation"])
plt.plot(df2021_2["EconomicSupportIndex"], model, color="red")
plt.show()
```

```python
correlation = round(st.pearsonr(df2021_2["EconomicSupportIndex"],df2021_2["inflation"])[0],2)
print(correlation)
```

```python

df2020_2 = df2020.dropna(subset=['unemployment', 'StringencyIndex'])

slope, intercept, r, p, err = st.linregress(df2020_2["StringencyIndex"],df2020_2["unemployment"])
model = slope * df2020_2["StringencyIndex"] + intercept

plt.scatter(df2020_2["StringencyIndex"],df2020_2["unemployment"])
plt.plot(df2020_2["StringencyIndex"], model, color="red")
plt.show()
```

```python
correlation = round(st.pearsonr(df2020_2["StringencyIndex"],df2020_2["unemployment"])[0],2)
print(correlation)
```
