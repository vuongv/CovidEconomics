```python
import pandas as pd
import os
```

```python
csv_path = r"C:\Users\liu12\Documents\Bootcamp\Covid_Economics\Resources\BalanceofTrade.csv"
bot_df = pd.read_csv(csv_path, encoding="latin1")

#Dropped first 4 rows
bot_df = bot_df.iloc[4:, :]
bot_df = bot_df.reset_index(drop=True)

# set the first row as the new header
new_header = bot_df.iloc[0] # grab the first row for the header
bot_df = bot_df[1:] # take the data excluding the header row
bot_df = bot_df.rename(columns = new_header) # set the header row as the DataFrame header

# drop the 'Base Year' column
bot_df = bot_df.drop('Base Year', axis=1)

bot_df = bot_df.rename(columns={'2017': 'bot_2017',
                        '2018': 'bot_2018',
                        '2019': 'bot_2019',
                        '2020': 'bot_2020',
                        '2021': 'bot_2021'})
bot_df = pd.melt(bot_df, id_vars=['Country'], value_vars=['bot_2017', 'bot_2018', 'bot_2019', 'bot_2020', 'bot_2021'], var_name='date', value_name='bot')

# select rows containing 'bot_2017' or 'exchange_2017' in the 'date' column
mask = (bot_df['date'] == 'bot_2017')
bot_df.loc[mask, 'date'] = '2017-01-01'  # replace with '2017'

mask = (bot_df['date'] == 'bot_2018')
bot_df.loc[mask, 'date'] = '2018-01-01'

mask = (bot_df['date'] == 'bot_2019')
bot_df.loc[mask, 'date'] = '2019-1-1'

mask = (bot_df['date'] == 'bot_2020')
bot_df.loc[mask, 'date'] = '2020-1-1'

mask = (bot_df['date'] == 'bot_2021')
bot_df.loc[mask, 'date'] = '2021-1-1'

bot_df = bot_df.sort_values(by=['Country', 'date'])

bot_df = bot_df.reset_index(drop=True)

bot_df.head(20)
```

```python
# merge the first two data frames
merged_df = pd.merge(bot_df, exchange_df, on=['Country','date'], how='outer')

# merge the third data frame
merged_df = pd.merge(merged_df, inflation_df, on=['Country','date'], how='outer')

# merge the fourth data frame
merged_df = pd.merge(merged_df, interest_df, on=['Country','date'], how='outer')

# merge the fifth data frame
merged_df = pd.merge(merged_df, liquidity_df, on=['Country','date'], how='outer')

# merge the sixth data frame
merged_df = pd.merge(merged_df, debt_df, on=['Country','date'], how='outer')
merged_df = merged_df.rename(columns={'Country': 'country'})
merged_df = merged_df.rename(columns={'bot_x': 'bot'})
merged_df = merged_df.drop('bot_y', axis=1)
merged_df["date"] = merged_df["date"].astype(str)


merged_df.head(50)
```
