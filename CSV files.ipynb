```python
import pandas as pd
import os
```

```python
csv_path = r"C:\Users\liu12\Documents\Bootcamp\Covid_Economics\Resources\BalanceofTrade.csv"
bot_df = pd.read_csv(csv_path, encoding="latin1")

#Dropped first 4 rows
bot_df = bot_df.iloc[4:, :]
bot_df = bot_df.reset_index(drop=True)

# set the first row as the new header
new_header = bot_df.iloc[0] # grab the first row for the header
bot_df = bot_df[1:] # take the data excluding the header row
bot_df = bot_df.rename(columns = new_header) # set the header row as the DataFrame header

# drop the 'Base Year' column
bot_df = bot_df.drop('Base Year', axis=1)

bot_df = bot_df.rename(columns={'2017': 'bot_2017',
                        '2018': 'bot_2018',
                        '2019': 'bot_2019',
                        '2020': 'bot_2020',
                        '2021': 'bot_2021'})

bot_df.head()
```

```python
csv_path_2 = r"C:\Users\liu12\Documents\Bootcamp\Covid_Economics\Resources\Exchange_Rates.csv"
exchange_df = pd.read_csv(csv_path_2, encoding="latin1")

#Dropped first 5 rows
exchange_df = exchange_df.iloc[5:, :]
exchange_df = exchange_df.reset_index(drop=True)

# set the first row as the new header
new_header = exchange_df.iloc[0] # grab the first row for the header
exchange_df = exchange_df[1:] # take the data excluding the header row
exchange_df = exchange_df.rename(columns = new_header) # set the header row as the DataFrame header

# drop the 'NaN Year' column
exchange_df = exchange_df.dropna(axis=1, how='all')

exchange_df = exchange_df.rename(columns={'2017': 'exchange_2017',
                        '2018': 'exchange_2018',
                        '2019': 'exchange_2019',
                        '2020': 'exchange_2020',
                        '2021': 'exchange_2021',
                        '2022':'exchange_2022'})

exchange_df.head()
```

```python
csv_path_3 = r"C:\Users\liu12\Documents\Bootcamp\Covid_Economics\Resources\Inflation_Prices.csv"
inflation_df = pd.read_csv(csv_path_3, encoding="latin1")

#Dropped first 4 rows
inflation_df = inflation_df.iloc[4:, :]
inflation_df = inflation_df.reset_index(drop=True)

# set the first row as the new header
new_header = inflation_df.iloc[0] # grab the first row for the header
inflation_df = inflation_df[1:] # take the data excluding the header row
inflation_df = inflation_df.rename(columns = new_header) # set the header row as the DataFrame header

# drop the 'NaN Year' column
inflation_df = inflation_df.dropna(axis=1, how='all')
inflation_df = inflation_df.drop('Base Year', axis=1)

inflation_df = inflation_df.rename(columns={'2017': 'inflation_2017',
                        '2018': 'inflation_2018',
                        '2019': 'inflation_2019',
                        '2020': 'inflation_2020',
                        '2021': 'inflation_2021',
                        '2022':'inflation_2022'})

inflation_df.head()
```

```python
csv_path_4 = r"C:\Users\liu12\Documents\Bootcamp\Covid_Economics\Resources\Interest_Rates.csv"
interest_df = pd.read_csv(csv_path_4, encoding="latin1")

#Dropped first 5 rows
interest_df = interest_df.iloc[5:, :]
interest_df = interest_df.reset_index(drop=True)

# set the first row as the new header
new_header = interest_df.iloc[0] # grab the first row for the header
interest_df = interest_df[1:] # take the data excluding the header row
interest_df = interest_df.rename(columns = new_header) # set the header row as the DataFrame header

# drop the 'NaN Year' column
interest_df = interest_df.dropna(axis=1, how='all')
interest_df = interest_df.rename(columns={2017.0: 'interest_2017',
                        '2018': 'interest_2018',
                        '2019': 'interest_2019',
                        '2020': 'interest_2020',
                        '2021': 'interest_2021',
                        '2022':'interest_2022'})


interest_df.head()
```

```python
csv_path_5 = r"C:\Users\liu12\Documents\Bootcamp\Covid_Economics\Resources\International_Liquidity.csv"
liquidity_df = pd.read_csv(csv_path_5, encoding="latin1")

#Dropped first 5 rows
liquidity_df = liquidity_df.iloc[5:, :]
liquidity_df = liquidity_df.reset_index(drop=True)

# set the first row as the new header
new_header = liquidity_df.iloc[0] # grab the first row for the header
liquidity_df = liquidity_df[1:] # take the data excluding the header row
liquidity_df = liquidity_df.rename(columns = new_header) # set the header row as the DataFrame header

# drop the 'NaN Year' column
liquidity_df = liquidity_df.dropna(axis=1, how='all')

liquidity_df = liquidity_df.rename(columns={'2017': 'liquidity_2017',
                        '2018': 'liquidity_2018',
                        '2019': 'liquidity_2019',
                        '2020': 'liquidity_2020',
                        '2021': 'liquidity_2021',
                        '2022':'liquidity_2022'})

liquidity_df.head()
```

```python
csv_path_6 = r"C:\Users\liu12\Documents\Bootcamp\Covid_Economics\Resources\Debt to GDP ratio.csv"
debt_df = pd.read_csv(csv_path_6, encoding="latin1")

#Dropped first 1 rows
debt_df = debt_df.iloc[1:, :]
debt_df = debt_df.reset_index(drop=True)

debt_df = debt_df.rename(columns={'2017': 'debt_2017',
                        '2018': 'debt_2018',
                        '2019': 'debt_2019',
                        '2020': 'debt_2020',
                        '2021': 'debt_2021',
                        '2022':'debt_2022',
                        'Central Government Debt (Percent of GDP)': 'Country'
                        })
debt_df.head()
```

```python
# merge the first two data frames
merged_df = pd.merge(bot_df, exchange_df, on=['Country','Scale'], how='outer')

# merge the third data frame
merged_df = pd.merge(merged_df, inflation_df, on=['Country','Scale'], how='outer')

# merge the fourth data frame
merged_df = pd.merge(merged_df, interest_df, on=['Country','Scale'], how='outer')

# merge the fifth data frame
merged_df = pd.merge(merged_df, liquidity_df, on=['Country','Scale'], how='outer')

# merge the sixth data frame
merged_df = pd.merge(merged_df, debt_df, on='Country', how='outer')

merged_df.head()
```
