```python
import pandas as pd
import numpy as np
```

```python
# Read the inflation.csv file into a dataframe
growth_df = pd.read_csv("Final_Complete.csv")
growth_df = growth_df.sort_values(['country', 'year'])
```

```python
for i in range(1, len(growth_df)):
    if growth_df.loc[i, 'country'] == growth_df.loc[i-1, 'country']:
        growth_df.loc[i, 'GDP_change'] = growth_df.loc[i, 'GDP'] - growth_df.loc[i-1, 'GDP']
growth_df = growth_df.drop('Unnamed: 0', axis=1)
growth_df = growth_df.reset_index(drop=True)
growth_df
```

```python
df1 = growth_df.loc[:, ['country', 'year','GDP']]
df1
```

```python

df_clean= df1.replace(0, np.nan)  # Replace zero values with NaN
df_clean = df_clean.dropna() # Remove rows containing NaN values

df_clean
```

```python
count_values = df_clean.groupby('country').size().reset_index(name='Count')

count_values
```

```python
gdp_growth = (df_clean.groupby('country')['GDP']
                .apply(lambda x: (((x.iloc[-1])-(x.iloc[0]))/x.iloc[0])*100)
                .reset_index(name='GDP growth'))

gdp_growth
```

```python
merged_df = pd.merge(gdp_growth, count_values, on='country', how='inner')

merged_df
```

```python
merged_df['Agg. GDP Growth'] = merged_df['GDP growth'] / merged_df['Count']


merged_df
```
