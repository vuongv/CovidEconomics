```python
import pandas as pd
import requests as re

country_names_df = pd.read_csv("countries.csv")
country_names_df = country_names_df.rename(columns = {"alpha_3": "ISO"})
country_names_df = country_names_df[["country", "ISO"]] 
```

# Population Data

```python
countries_iso = country_names_df["ISO"]

date = "2018:2021"
indicator_string = "SP.POP.TOTL"

year = ""
value = ""
test = []
results = []

for x in countries_iso:
    
    data = re.get(f"http://api.worldbank.org/v2/country/{x}/indicator/{indicator_string}?date={date}&format=json").json()

    try:
        entries = len(data[1])
        for a in range(entries):
        
            row_value = data[1][a]["value"]
            row_date = data[1][a]["date"]
            new_row = {"ISO": x, "year": row_date, "population": row_value}
            results.append(new_row)
    except:
        print(f"Skipped {x}")

print("Complete")

population_df = pd.DataFrame(results)
```

```python
working_data_df = pd.merge(country_names_df ,population_df, on = ["ISO"])
working_data_df

countries_iso_clean = working_data_df["ISO"].unique()
```

# Inflation Data

```python
countries_iso = country_names_df["ISO"]

date = "2018:2021"
indicator_string = "FP.CPI.TOTL.ZG"

year = ""
value = ""
test = []
results = []

for x in countries_iso_clean:
    
    data = re.get(f"http://api.worldbank.org/v2/country/{x}/indicator/{indicator_string}?date={date}&format=json").json()

    try:
        entries = len(data[1])
        for a in range(entries):
        
            row_value = data[1][a]["value"]
            row_date = data[1][a]["date"]
            new_row = {"ISO": x, "year": row_date, "inflation": row_value}
            results.append(new_row)
    except:
        print(f"Skipped {x}")

print("Complete")

inflation_df = pd.DataFrame(results)
```

# GDP Data

```python
date = "2018:2021"
indicator_string = "NY.GDP.MKTP.CD"

year = ""
value = ""
test = []
results = []

for x in countries_iso_clean:
    
    data = re.get(f"http://api.worldbank.org/v2/country/{x}/indicator/{indicator_string}?date={date}&format=json").json()

    try:
        entries = len(data[1])
        for a in range(entries):
        
            row_value = data[1][a]["value"]
            row_date = data[1][a]["date"]
            new_row = {"ISO": x, "year": row_date, "GDP": row_value}
            results.append(new_row)
    except:
        print(f"Skipped {x}")

print("Complete")

GDP_df = pd.DataFrame(results)
```

# Unemployment Data

```python
date = "2018:2021"
indicator_string = "SL.UEM.TOTL.NE.ZS"

year = ""
value = ""
test = []
results = []

for x in countries_iso_clean:
    
    data = re.get(f"http://api.worldbank.org/v2/country/{x}/indicator/{indicator_string}?date={date}&format=json").json()

    try:
        entries = len(data[1])
        for a in range(entries):
        
            row_value = data[1][a]["value"]
            row_date = data[1][a]["date"]
            new_row = {"ISO": x, "year": row_date, "unemployment": row_value}
            results.append(new_row)
    except:
        print(f"Skipped {x}")

print("Complete")

unemployment_df = pd.DataFrame(results)
```

```python
unemployment_df
```

# Combine

```python
working_data_df = pd.merge(working_data_df ,GDP_df, on = ["ISO", "year"])
working_data_df = pd.merge(working_data_df ,inflation_df, on = ["ISO", "year"])
working_data_df = pd.merge(working_data_df ,unemployment_df, on = ["ISO", "year"])
working_data_df
```

```python
working_data_df.to_csv("Working Data.csv")
```

# Get Clean Data

```python
clean_data_df = working_data_df
for x in countries_iso_clean:
    if clean_data_df[clean_data_df["ISO"] == x].isnull().any().any():
        clean_data_df = clean_data_df[clean_data_df["ISO"] != x]

clean_data_df = clean_data_df.reset_index(drop=True)
clean_data_df
```

```python
clean_data_df.to_csv("Clean Data.csv")
```

# GDP Growth

```python
results = []

for ISO in clean_data_df["ISO"].unique():
    
    start = clean_data_df[(clean_data_df["ISO"] == ISO) & (clean_data_df["year"] == "2020")]["GDP"].values[0]
    end = clean_data_df[(clean_data_df["ISO"] == ISO) & (clean_data_df["year"] == "2021")]["GDP"].values[0]

    change = ((start - end) / start) * 100
    
    new_row = {"ISO": ISO,"2020": start, "2021": end,  "change": change}

    results.append(new_row)

GDP_change_df = pd.DataFrame(results)
GDP_change_df
```

```python
GDP_change_df.to_csv("GDP Change.csv")
```
